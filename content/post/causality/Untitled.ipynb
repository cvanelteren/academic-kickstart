```python
import numpy as np, matplotlib.pyplot as plt

from scipy import integrate
def update(t, x):
    x1, x2, x3  = x
    dx = sigma * (x2 - x1)
    dy = x1 * (rho - x3) - x2
    dz = x1 * x2 - beta * x3
    return x + np.asarray([dx, dy,  dz])


y0 = np.asarray([0, 0, 0])
y0 = np.random.rand(3) 
t = np.linspace(0, 100, 1000)
sigma = 10
rho = 28
beta = 8/3

# m = integrate.odeint(update, y0, t)
m = integrate.solve_ivp(update, [0,100], y0, vectorized = 1,\
                       method = 'LSODA')
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
fig, ax = plt.subplots(subplot_kw = dict(projection = '3d'))
ax.plot(*m.y, alpha = .5, linestyle = 'dashed')


fig.show()
```
